```python
import torch
import pandas
import numpy
```

```python
data = pandas.read_csv("CalibratorTestingMultipleFiles.csv")
data
```

```python
vocabularyDataframe = pandas.read_csv("VocabularyForTransformerUnimod_V4.csv", index_col="Token")
vocab_dict = vocabularyDataframe.to_dict()["Id"]
vocab_dict["M"]
```

```python
dataset = data[["FullSequence", "Mean"]]
dataset
```

```python
for index, row in dataset.iterrows():
    # print(row[0])
    break


import re as regex
pre_tokens = []
tokens = []
for index, row in dataset.iterrows():
    sequence = str(row[0])
    if(sequence.count("[") > 0):
        stars = regex.sub( "(?<=[A-HJ-Z])\\[|(?<=\\[)[A-HJ-Z](?=\\])|(?<=[A-HJ-Z])\\](?=$|[A-Z]|(?<=\\])[^A-Z])",
                    "*", sequence)
        removedColon = regex.sub("\\*(.*?):", "*", stars)
        pre_tokens.append((removedColon, row[1]))
    else:
        pre_tokens.append((sequence, row[1]))
```

```python
len(pre_tokens)
```

```python
tokens = []
for sequence in pre_tokens:
    tokenList = []
    for subSequence in sequence[0].split("*"):
        if("on" not in subSequence):
            for residue in subSequence:
                if(residue in vocab_dict):
                    tokenList.append(vocab_dict[residue])
                else:
                    tokenList.clear()
                    break
        else:
            if(subSequence in vocab_dict):
                tokenList.append(vocab_dict[subSequence])
            else:
                tokenList.clear()
                break
    if(len(tokenList) != 0):
        while(len(tokenList) != 100):
            tokenList.append(0)
        tokens.append((tokenList, sequence[1]))
```

```python
trainingSequencesAsArray = numpy.array(trainingDataFrame[0].values.tolist(), dtype="int")
trainingRetentionTimesAsArray = numpy.array(trainingDataFrame[1].values.tolist(), dtype="float32")
testingSequencesAsArray = numpy.array(testingDataFrame[0].values.tolist(), dtype="int")
testingRetentionTimesAsArray = numpy.array(testingDataFrame[1].values.tolist(), dtype="float32")
```

```python
trainingSequencesAsArray, trainingRetentionTimesAsArray, trainingSequencesAsArray.shape, trainingRetentionTimesAsArray.shape
```

```python
trainingFeatures = torch.tensor(trainingSequencesAsArray, dtype=torch.long)
trainingLabels = torch.tensor(trainingRetentionTimesAsArray, dtype=torch.float32)
testingFeatures = torch.tensor(testingSequencesAsArray, dtype=torch.long)
testingLabels = torch.tensor(testingRetentionTimesAsArray, dtype=torch.float32)
```

```python
trainingFeatures, trainingLabels
```

```python
# Define the model architecture
class AttentionRegression(torch.nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(AttentionRegression, self).__init__()
        self.embedding = torch.nn.Embedding(input_size, 32)
        self.lstm = torch.nn.LSTM(32, hidden_size, batch_first=True)
        self.attention = torch.nn.MultiheadAttention(hidden_size, num_heads=1)
        self.fc = torch.nn.Linear(hidden_size, output_size)
        self.sigmoid = torch.nn.Sigmoid()

    def forward(self, x):
        embedded = self.embedding(x)
        lstm_out, _ = self.lstm(embedded)
        lstm_out = lstm_out.permute(1, 0, 2)  # [seq_len, batch, hidden_size]
        attention_output, _ = self.attention(lstm_out, lstm_out, lstm_out)
        output = self.fc(attention_output.mean(dim=0))
        output = self.sigmoid(output)
        return output
#Dataset 
class PeptidesWithRetentionTimes(torch.utils.data.Dataset):
    def __init__(self, peptides, retentionTime):
        self.peptides = peptides
        self.retentionTimes = retentionTime

    def __len__(self):
        return len(self.peptides)

    def __getitem__(self, index):
        peptide = self.peptides[index]
        retentionTime = self.retentionTimes[index]
        return peptide, retentionTime
```

```python
trainingTorchDataset = PeptidesWithRetentionTimes(trainingFeatures, trainingLabels)
trainingDataloader = torch.utils.data.DataLoader(trainingTorchDataset, batch_size=64, shuffle=True, drop_last=True)
testingTorchDataset = PeptidesWithRetentionTimes(testingFeatures, testingLabels)
testingDataloader = torch.utils.data.DataLoader(testingTorchDataset, batch_size=64, shuffle=True, drop_last=True)
```
