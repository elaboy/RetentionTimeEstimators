```python
import pandas 
import glob
```

```python
path = "C:\\Users\elabo\Documents\RetentionFileDatasets"
all_files = glob.glob(path + "/*.csv")
```

```python
dataframes = []
for filename in all_files:
    df = pandas.read_csv(filename, index_col=None, header=0)
    dataframes.append(df)
```

```python
dataframes
```

```python
df = pandas.concat(dataframes, axis=0, ignore_index=True)
df.to_csv('RetentionFileDatasets.csv', index=False)
```

```python
pandas.plotting.scatter_matrix(df, alpha=0.2, figsize=(6, 6), diagonal='kde')
```
